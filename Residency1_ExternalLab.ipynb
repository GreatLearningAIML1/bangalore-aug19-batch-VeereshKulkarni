### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
# 75th Percentile
Q3 = df['Age'].quantile(0.75)
# 25th Percentile
Q1 = df['Age'].quantile(0.25)
print(Q3)
print(Q1)
print("IQR - Interquartile range =",Q3-Q1)
```
